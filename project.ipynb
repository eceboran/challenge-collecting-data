```python
url = "https://www.immoweb.be/en/classified/house/for-sale/mortsel/2640/9529029?searchId=616432a98f262"
r = requests.get(url)
print(url, r.status_code)
#saving to txt file
print(type(r.text))
text_file = open("content.txt", "w", encoding='utf-8')
n = text_file.write(r.text)
text_file.close()
```

```python
#reading txt file with bs
soup = BeautifulSoup(open("content.txt"))
print(soup.prettify())
```

```python
script = soup.find_all('script')[0].string.strip()
script = script[20:-2].strip()
data = json.loads(script)
print(data)
```

```python
classified = data['classified']

house_id = classified['id']
house_type = classified['type']
subtype = classified['subtype']
price = classified['price']
transaction_type = classified['transactionType']
kitchen_type = classified['kitchen']['type'] # if 'equipped' then 'yes' ??
room_number = classified['bedroom']['count']
area = classified['land']['surface']
garden = classified['outdoor']['garden']['surface']
terrace = classified['outdoor']['terrace']['exists']
swimming_pool = classified['wellnessEquipment']['hasSwimmingPool']
state_of_the_building = classified['condition']['isNewlyBuilt']

print(house_id)
print(house_type)
print(subtype)
print(price)
print(transaction_type)
print(kitchen_type)
print(room_number)
print(area)
print(garden)
print(terrace)
print(swimming_pool)
print(state_of_the_building)
```

```python
house_ids = []
house_types = []
subtypes = []
prices = []
transaction_types = []
kitchen_types = []
room_numbers = []
areas = []
gardens = []
terraces = []
swimming_pools = []
state_of_the_buildings = []


house_ids.append(house_id)
house_types.append(house_type)
subtypes.append(subtype)
prices.append(price)
transaction_types.append(transaction_type)
kitchen_types.append(kitchen_type)
room_numbers.append(room_number)
areas.append(area)
gardens.append(garden)
terraces.append(terrace)
swimming_pools.append(swimming_pool)
state_of_the_buildings.append(state_of_the_building)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"House Id": house_ids})
df["House Type"] = house_types
df["Subtype"] = subtypes
df["Price"] = prices
df["Transaction Type"] = transaction_types
df["Kitchen Type"] = kitchen_types
df["Room Number"] = room_numbers
df["Area"] = areas
df["Garden"] = gardens
df["Terrace"] = terraces
df["Swimming Pool"] = swimming_pools
df["State of Building"] = state_of_the_buildings

```

```python
print(df.values)
```

```python
df.to_csv("./real_estate_offers.csv", index=False)
```

```python
driver = webdriver.Firefox()
url = "https://www.immoweb.be/en/classified/house/for-sale/mortsel/2640/9529029?searchId=616432a98f262"
driver.get(url)
elements = driver.find_elements_by_class_name('classified__information--address-row')
street_element = elements[0]
postcode_locality = elements[1]
locality_elements  = postcode_locality.find_elements_by_tag_name('span')
postcode = locality_elements[0].text
locality = locality_elements[-1].text
print(postcode)
print(locality)

immoweb_code_class= driver.find_elements_by_class_name('classified__header--immoweb-code')

for element in immoweb_code_class:
    immoweb_code = element.text
    print(immoweb_code)


driver.close()

```
