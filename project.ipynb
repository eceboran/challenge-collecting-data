```python
# Url of website
url = "https://www.immoweb.be/en/search/house-and-apartment/for-sale?countries=BE&page=1&orderBy=relevance"
```

```python
# Use request to get the soup
# I send my HTTP request with a "GET" to the site server to identify in the url
r = requests.get(url)
# I display the requested url and the return of the server
print(url, r.status_code)
# I ask beautifulsoup to store the HTML content of the website in the `soup` variable
soup_request = BeautifulSoup(r.content, "html.parser")
soup_request.prettify()
print(soup_request)
```

```python
# This one doesn't contain the links
soup_request.find_all("a", attrs={"class": "card__title-link"})
```

```python
# Use the driver to open the website
driver = webdriver.Firefox()
driver.implicitly_wait(10)
driver.get(url)
```

```python
# BeautifuLSoup using the driver
soup_driver = BeautifulSoup(driver.page_source)
print(soup_driver)
```

```python
links_driver = []
for value in driver.find_elements_by_xpath('//li//article//div//h2//a[@class="card__title-link"]'):
    print(value.get_attribute("href"))
    links_driver.append(value.get_attribute("href"))
print(len(links_driver))
print(type(links_driver))
```

```python
# Also with BeautifulSoup
links_soup = []
for elem in soup_driver.find_all("a", attrs={"class": "card__title-link"}):
    print(elem.get("href"))
    links_soup.append(elem.get("href"))
links_soup
print(type(links_soup[0]))
```

```python
# Also with BeautifulSoup
links_soup = []
for elem in soup_driver.select("li.search-results__item a.card__title-link"):
    print(elem.get("href"))
    links_soup.append(elem.get("href"))
links_soup
print(type(links_soup[0]))
```

```python
# Compare the links from the driver and from Beautiful Soup
for index, elem in enumerate(links_driver):
    if index<3:
        print(elem,' ',links_soup[index])
    print(elem == links_soup[index])
```

```python
# Close the driver
driver.close()
```

```python
# Use request to get the soup
# I send my HTTP request with a "GET" to the site server to identify in the url
r = requests.get(url)
# I display the requested url and the return of the server
print(url, r.status_code)
# I ask beautifulsoup to store the HTML content of the website in the `soup` variable
soup_request = BeautifulSoup(r.content, "html.parser")
soup_request.prettify()
print(soup_request)
```

```python
# Also with BeautifulSoup
links_req = []
for elem in soup_request.select("p.classified__information--address-clickable span.classified__information--address-row"):
    print(elem.text)
    print('print next')
    links_req.append(elem.text)
print(len(links_req))
print(links_req)
```

```python
# Use the driver to open the website
driver = webdriver.Firefox()
driver.implicitly_wait(10)
driver.get(url)
```

```python
# BeautifuLSoup using the driver
soup_driver = BeautifulSoup(driver.page_source)
print(soup_driver)
```

```python
links_driver = []
for value in driver.find_elements_by_xpath('//p[@class="classified__information--address-clickable"]'):
    print(len(value.find_elements_by_tag_name('span')))
    for value_2 in value.find_elements_by_tag_name('span'):
        print(value_2.text)
        links_driver.append(value_2.text)
    print(value.text)
print(len(links_driver))
print(type(links_driver))
print(links_driver)
```

```python
links_driver = []
address_tags = driver.find_elements_by_xpath('//p[@class="classified__information--address-clickable"]//span[@class="classified__information--address-row"]')
print(len(address_tags))
for value in address_tags:
    links_driver.append(value.text)
    #print(value.text)
print(len(links_driver))
print(type(links_driver))
print(links_driver)
```

```python
# Also with BeautifulSoup
links_soup = []
for elem in soup_driver.select("p.classified__information--address-clickable span.classified__information--address-row"):
    print(elem.text)
    print('print next')
    links_soup.append(elem.text)
print(len(links_soup))
print(links_soup)
```

```python
# Also with BeautifulSoup
links_soup = []
print(len(soup_driver.find_all("p", attrs={"class": "classified__information--address-clickable"})))
for elem in soup_driver.find_all("p", attrs={"class": "classified__information--address-clickable"}):
    # print(type(elem.contents[0]))
    # print(elem)
    print(len(elem.find_all("span", attrs={"class": "classified__information--address-row"})))
    for elem_2 in elem.find_all("span", attrs={"class": "classified__information--address-row"}):
        print(elem_2.text)
        links_soup.append(elem_2.text)
print(links_soup)
```
