## Save listing contents as text
Save the contents of each listing (house or apartment) into a text file, to be used later by BeautifulSoup.

```python
# Import necessary libraries
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
import random
import re
from queue import Queue
from threading import Thread
import requests
import json
import pandas as pd
```

```python
# Create a folder for data
main_path = os.path.abspath("")
data_folder_path = os.path.join(main_path,"data/")
if not os.path.exists(data_folder_path):
    os.makedirs(data_folder_path)
print(main_path)
print(data_folder_path)
```

```python
# Folder for page contents
contents_folder_path = os.path.join(data_folder_path,"contents/")
if not os.path.exists(contents_folder_path):
    os.makedirs(contents_folder_path)
```

```python
print(os.listdir(contents_folder_path))
```

```python
# Get the list of files
file_list_raw = os.listdir(contents_folder_path)
file_name_list = []
file_path_list = []
# print(file_list)
for file_name in file_list_raw:
    m = re.match("Contents_[0-9]+.txt", file_name)
    if m is not None:
        file_name_list.append(file_name)
        file_path_list.append(os.path.join(contents_folder_path,file_name))
    else:
        print("Ignored: " + file_name)
print(len(file_list_raw))
print(len(file_name_list))
print(len(file_path_list))
```

```python
#soup = BeautifulSoup(open(file_path_list[0]))
#print(soup.prettify())
```

```python
# Run a loop over files
house_ids = []
localities = []
house_types = []
subtypes = []
prices = []
transaction_types = []
kitchen_types = []
room_numbers = []
areas = []
gardens = []
terraces = []
swimming_pools = []
state_of_the_buildings = []

# Load each file
for no_file, file_name in enumerate(file_name_list):
    
    # File path
    file_path = file_path_list[no_file]
    
    print(file_name)
    print(file_path)
    
    # Immoweb code from the file name
    m = re.search("Contents_(?P<immoweb_code>\d+).txt$", file_name)
    immoweb_code = m.group("immoweb_code")
    print(immoweb_code)
    print(f"Running file {no_file}")
        
    # Reading txt file with bs
    with open(file_path, 'r', encoding="utf8") as f:
        soup = BeautifulSoup(f.read())

    script = soup.find_all('script')[0].string.strip()
    script = script[20:-2].strip()
    data = json.loads(script)
    # print(data)
    
    if "classified" not in data.keys():
        print("Page not available anymore")
        continue
        
    classified = data['classified']

    house_id = classified['id']
    locality = classified['zip']
    house_type = classified['type']
    subtype = classified['subtype']
    price = classified['price']
    # transaction_type = classified['transactionType']
    kitchen_type = classified['kitchen']['type'] # if 'equipped' then 'yes' ??
    room_number = classified['bedroom']['count']
    area = classified['land']['surface']
    garden = classified['outdoor']['garden']['surface']
    terrace = classified['outdoor']['terrace']['exists']
    swimming_pool = classified['wellnessEquipment']['hasSwimmingPool']
    state_of_the_building = classified['condition']['isNewlyBuilt']
    
    # Retrieve information using beautiful soup

    # Save information in a variable
    house_ids.append(house_id)
    localities.append(locality)
    house_types.append(house_type)
    subtypes.append(subtype)
    prices.append(price)
    # transaction_types.append(transaction_type)
    kitchen_types.append(kitchen_type)
    room_numbers.append(room_number)
    areas.append(area)
    gardens.append(garden)
    terraces.append(terrace)
    swimming_pools.append(swimming_pool)
    state_of_the_buildings.append(state_of_the_building)
```

```python
df = pd.DataFrame({"House Id": house_ids})
df["Locality"] = localities
df["House Type"] = house_types
df["Subtype"] = subtypes
df["Price"] = prices
# df["Transaction Type"] = transaction_types
df["Kitchen Type"] = kitchen_types
df["Room Number"] = room_numbers
df["Area"] = areas
df["Garden"] = gardens
df["Terrace"] = terraces
df["Swimming Pool"] = swimming_pools
df["New Building"] = state_of_the_buildings

print(df.values)
```

```python
csv_raw_file_name = "real_estate_offers_raw.csv"
csv_raw_file_path = os.path.join(main_path,csv_raw_file_name)
```

```python
# Save csv file
df.to_csv(csv_raw_file_path, index=False)
```
