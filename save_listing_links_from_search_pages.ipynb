## Save listing links from search pages
Save the raw and processed links of each listing (house or apartment) into text files.

```python
# Import necessary libraries
from selenium import webdriver
from bs4 import BeautifulSoup
import os
import time
import random
import re
```

```python
# Create a folder for data
data_folder_path = os.path.abspath("./data/")
if not os.path.exists(data_folder_path):
    os.makedirs(data_folder_path)
```

```python
# Create the txt file for the links of the individual listings
listings_url_file_name = "listings_url_list_raw.txt" 
listings_url_file_path = os.path.join(data_folder_path,listings_url_file_name)
if os.path.exists(listings_url_file_path) is not True:
    open(listings_url_file_path, 'w').close()
```

```python
# Start the driver to open the website
driver = webdriver.Firefox()
driver.implicitly_wait(10)
```

```python
# Retrieve links for listings in each search results page, as long as there are more results
no_search_result_page = 1
url_list_all_pages = []
while True:
    url_search_result = f"https://www.immoweb.be/en/search/house-and-apartment/for-sale?countries=BE&page={no_search_result_page}&orderBy=relevance"
    print(no_search_result_page)
    no_search_result_page += 1
    # Go to the current search results webpage
    time.sleep(random.uniform(3.0, 8.0))  # wait
    driver.get(url_search_result)
    # Get BeautifulSoup using the driver
    soup = BeautifulSoup(driver.page_source)
    # Close the driver
    # Get the urls
    url_list_single_page = []
    for elem in soup.select("li.search-results__item a.card__title-link"):
        href_listing = elem.get("href")
        index_search = href_listing.find("?searchId")
        if index_search != -1:
            href_listing = href_listing[:index_search]
        url_list_single_page.append(href_listing)
    if len(url_list_single_page) == 0:
        break
    # Add the urls to the list for all search pages
    url_list_all_pages.extend(url_list_single_page)
```

```python
# Close the driver
driver.close()
```

```python
print(len(url_list_all_pages))
```

```python
# Save the raw links to the text file
with open(listings_url_file_path, "r+") as file:
    file.seek(0, 2)
    for url in url_list_all_pages:
        file.writelines(url + '\n')
```

Load the links from the file for further processing

```python
# Load the contents of the raw links file for cleaning
with open(listings_url_file_path, "r") as file:
    url_list_raw = file.readlines()
```

```python
# Remove repeated entries
url_list_processed = list(set(url_list_raw))
print(len(url_list_processed))
```

```python
# Number of links for each type of project
listing_types = []
indices_to_delete = []
for index, url in enumerate(url_list_processed):
    m = re.search(
    "https://www.immoweb.be/en/classified/(?P<listing_type>[\w\-]+)/for-sale/",
    url)
    listing_type = m.group("listing_type") 
    listing_types.append(listing_type)
    if (listing_type == "new-real-estate-project-houses") or (listing_type == "new-real-estate-project-apartments"):
        indices_to_delete.append(index)
print([[x,listing_types.count(x)] for x in set(listing_types)])
```

```python
print(len(indices_to_delete))
print(indices_to_delete)
```

```python
# Delete new real estate projects from the list
for index in indices_to_delete:
    print(url_list_processed[index])
    url_list_processed[index] = ''
```

```python
url_list_processed = list(set(url_list_processed))
url_list_processed.remove('')
print(url_list_processed)
```

```python
# Create the txt file for the links of the individual listings
listings_url_file_name_2 = "listings_url_list_processed.txt" 
listings_url_file_path_2 = os.path.join(data_folder_path,listings_url_file_name_2)
if os.path.exists(listings_url_file_path_2) is not True:
    open(listings_url_file_path_2, 'w').close()
```

```python
# Save the processed links to the text file
with open(listings_url_file_path_2, "w") as file:
    file.seek(0)
    for url in url_list_processed:
        file.writelines(url)
```
